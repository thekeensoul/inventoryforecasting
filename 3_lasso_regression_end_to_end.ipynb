# Lasso Regression — End-to-End

This notebook builds on the earlier pipelines and adds a Lasso Regression implementation with hyperparameter tuning (GridSearchCV). Assumes `retail_store_inventory.csv` is in the same directory as the notebook.

```python
# Imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
from sklearn.model_selection import train_test_split, GridSearchCV, TimeSeriesSplit
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
import joblib
%matplotlib inline

from google.colab import drive
drive.mount('/content/drive')
```

```python
# -------------------------
# Helper functions
# -------------------------
def smape(actual, forecast):
    """Symmetric Mean Absolute Percentage Error (avoids divide-by-zero)."""
    actual = np.array(actual, dtype=float)
    forecast = np.array(forecast, dtype=float)
    denom = (np.abs(actual) + np.abs(forecast))
    mask = denom == 0
    denom[mask] = 1.0
    return 100.0 * np.mean(2.0 * np.abs(forecast - actual) / denom)

def print_metrics(y_true, y_pred):
    mae = mean_absolute_error(y_true, y_pred)
    rmse = np.sqrt(mean_squared_error(y_true, y_pred))
    r2 = r2_score(y_true, y_pred)
    smape_val = smape(y_true, y_pred)
    print(f"MAE:  {mae:.4f}")
    print(f"RMSE: {rmse:.4f}")
    print(f"sMAPE: {smape_val:.4f}%")
    print(f"R²:   {r2:.4f}")
    return {'MAE': mae, 'RMSE': rmse, 'sMAPE': smape_val, 'R2': r2}
```

```python
# -------------------------
# Load dataset
# -------------------------
file_path = '/content/drive/MyDrive/data/retail_store_inventory.csv'
if not os.path.exists(file_path):
    raise FileNotFoundError(f"File not found: {file_path} - put 'retail_store_inventory.csv' in current directory")

df = pd.read_csv(file_path)
print("Loaded dataset shape:", df.shape)
display(df.head())
```

```python
# -------------------------
# Basic inspection & cleaning
# -------------------------
print("\nData types:")
print(df.dtypes)
print("\nMissing values per column:")
print(df.isnull().sum())

# Simple missing-value strategy: forward/backward fill (adjust as needed)
df = df.fillna(method='ffill').fillna(method='bfill')
```

```python
# -------------------------
# Feature engineering (date + simple)
# -------------------------
if 'Date' in df.columns:
    df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
    if df['Date'].isnull().any():
        df = df[df['Date'].notnull()].copy()
    df['Year'] = df['Date'].dt.year
    df['Month'] = df['Date'].dt.month
    df['Day'] = df['Date'].dt.day
    df['DayOfWeek'] = df['Date'].dt.dayofweek
    df['WeekOfYear'] = df['Date'].dt.isocalendar().week.astype(int)
    df['IsWeekend'] = df['DayOfWeek'].isin([5,6]).astype(int)

# Basic numeric sanity: ensure target exists
target_col = 'Units Sold'
if target_col not in df.columns:
    raise ValueError(f"Target column '{target_col}' not found. Available columns: {df.columns.tolist()}")

# Drop identifiers that are not predictive (if present)
drop_cols = []
for c in ['Store ID', 'Product ID', 'Date']:
    if c in df.columns:
        drop_cols.append(c)
if drop_cols:
    df = df.drop(columns=drop_cols)

print('\nAfter feature engineering, shape:', df.shape)
```

```python
# -------------------------
# Encode categorical variables (one-hot)
# -------------------------
cat_cols = df.select_dtypes(include=['object', 'category']).columns.tolist()
print("\nCategorical columns to encode:", cat_cols)
df = pd.get_dummies(df, columns=cat_cols, drop_first=True)
```

```python
# -------------------------
# Train/test split
# -------------------------
X = df.drop(columns=[target_col])
y = df[target_col]

# Quick check: if any infinite or NaN values in X, drop those rows
mask_valid = np.isfinite(X.select_dtypes(include=[np.number])).all(axis=1)
if not mask_valid.all():
    X = X[mask_valid].copy()
    y = y[mask_valid].copy()

# For time-series tasks, consider chronological split; here we do a random split for demonstration
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.20, random_state=42, shuffle=True
)

print("\nTrain shape:", X_train.shape, "Test shape:", X_test.shape)
```

```python
# -------------------------
# Preprocessing setup
# -------------------------
numeric_cols = X_train.select_dtypes(include=[np.number]).columns.tolist()
print('Numeric columns count:', len(numeric_cols))

preprocessor = ColumnTransformer(transformers=[('num', StandardScaler(), numeric_cols)], remainder='passthrough')
```

```python
# -------------------------
# Baseline: Linear Regression pipeline
# -------------------------
linear_pipe = Pipeline([
    ('preproc', preprocessor),
    ('lr', LinearRegression())
])

linear_pipe.fit(X_train, y_train)
y_pred_lr = linear_pipe.predict(X_test)
print('\nLinear Regression performance:')
metrics_lr = print_metrics(y_test, y_pred_lr)
```

```python
# -------------------------
# Ridge Regression (as a reference)
# -------------------------
ridge_pipe = Pipeline([
    ('preproc', preprocessor),
    ('ridge', Ridge(alpha=1.0))
])
ridge_pipe.fit(X_train, y_train)
y_pred_ridge_simple = ridge_pipe.predict(X_test)
print('\nRidge (alpha=1.0) performance:')
metrics_ridge_simple = print_metrics(y_test, y_pred_ridge_simple)
```

```python
# -------------------------
# Lasso Regression with GridSearchCV (hyperparameter tuning)
# -------------------------
lasso_pipe = Pipeline([
    ('preproc', preprocessor),
    ('lasso', Lasso(max_iter=10000))
])

param_grid = {
    'lasso__alpha': [0.0001, 0.001, 0.01, 0.1, 1.0, 10.0],
    'lasso__selection': ['cyclic', 'random']
}

# Use 5-fold CV; consider TimeSeriesSplit if chronological.
grid_lasso = GridSearchCV(lasso_pipe, param_grid, cv=5, scoring='neg_mean_squared_error', n_jobs=-1, verbose=1)
grid_lasso.fit(X_train, y_train)

print('\nBest Lasso params:', grid_lasso.best_params_)
print('Best CV score (neg MSE):', grid_lasso.best_score_)
```

```python
# -------------------------
# Evaluate best Lasso model on test set
# -------------------------
best_lasso = grid_lasso.best_estimator_
y_pred_lasso = best_lasso.predict(X_test)

print('\nLasso Regression performance on test set:')
metrics_lasso = print_metrics(y_test, y_pred_lasso)

# Compare with Linear and Ridge simple
import pandas as pd
compare = pd.DataFrame([metrics_lr, metrics_ridge_simple, metrics_lasso], index=['Linear', 'Ridge(alpha=1)', 'Lasso']).round(4)
display(compare)
```

```python
# -------------------------
# Coefficients (Lasso) - mapping back to original feature names
# -------------------------
lasso_step = best_lasso.named_steps['lasso']

# Reconstruct feature names order after ColumnTransformer + passthrough
remainder_cols = [c for c in X_train.columns if c not in numeric_cols]
feature_names = numeric_cols + remainder_cols

coef_df = pd.DataFrame({
    'feature': feature_names,
    'coefficient': lasso_step.coef_
})
coef_df['abs_coef'] = coef_df['coefficient'].abs()
coef_df = coef_df.sort_values(by='abs_coef', ascending=False).reset_index(drop=True)
print('\nTop 15 Lasso features by absolute coefficient:')
display(coef_df.head(15)[['feature','coefficient']])

# Show how many features were set to zero
zero_count = (coef_df['coefficient'] == 0).sum()
print(f"\nNumber of features with coefficient exactly zero: {zero_count} / {len(coef_df)}")
```

```python
# -------------------------
# Diagnostic plots for Lasso
# -------------------------
residuals_lasso = y_test - y_pred_lasso

plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
plt.scatter(y_pred_lasso, y_test, alpha=0.5)
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], '--', linewidth=1)
plt.xlabel('Predicted Units Sold')
plt.ylabel('Actual Units Sold')
plt.title('Lasso: Predicted vs Actual')

plt.subplot(1,2,2)
plt.hist(residuals_lasso, bins=40)
plt.xlabel('Residual (Actual - Predicted)')
plt.title('Lasso: Residuals Distribution')

plt.tight_layout()
plt.show()

plt.figure(figsize=(6,4))
plt.scatter(y_pred_lasso, residuals_lasso, alpha=0.4)
plt.axhline(0, linestyle='--', linewidth=1)
plt.xlabel('Predicted')
plt.ylabel('Residuals')
plt.title('Lasso: Residuals vs Predicted')
plt.show()
```

```python
# -------------------------
# Save best Lasso model and preprocessor
# -------------------------
save_artifacts = True
if save_artifacts:
    joblib.dump(best_lasso, 'best_lasso_pipeline.joblib')
    print('\nSaved artifact: best_lasso_pipeline.joblib')
```
